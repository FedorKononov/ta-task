````python
#Define the functions that create and train a model.

import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import feature_column
from tensorflow.keras import layers
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split

tf.compat.v1.logging.set_verbosity(tf.compat.v1.logging.ERROR)

def create_model(learning_rate, feature_layer, metrics):
  """Create and compile a simple classification model."""
  # Most simple tf.keras models are sequential.
  model = tf.keras.models.Sequential()

  # Add the feature layer (the list of features and how they are represented)
  # to the model.
  model.add(feature_layer)

  # Define the first hidden layer with 20 nodes.   
  model.add(tf.keras.layers.Dense(units=24, 
                                  activation='relu', 
                                  name='Hidden1'))
  
  # Define the second hidden layer with 12 nodes. 
  model.add(tf.keras.layers.Dense(units=12, 
                                  activation='relu', 
                                  name='Hidden2'))

  # Funnel the regression value through a sigmoid function.
  model.add(tf.keras.layers.Dense(units=1, input_shape=(1,),
                                  activation=tf.sigmoid),)

  # Call the compile method to construct the layers into a model that
  # TensorFlow can execute.
  model.compile(optimizer=tf.keras.optimizers.Adam(lr=learning_rate),                                                   
                loss=tf.keras.losses.BinaryCrossentropy(),
                metrics=metrics)

  return model        


def train_model(model, dataset, epochs, label_name,
                batch_size=None, shuffle=False):
  """Feed a dataset into the model in order to train it."""

  # The x parameter of tf.keras.Model.fit can be a list of arrays, where
  # each array contains the data for one feature.  Here, we're passing
  # every column in the dataset. The feature_layer will filter
  # away most of those columns, leaving only the desired columns and their
  # representations as features.
  features = {name:np.array(value) for name, value in dataset.items()}
  label = np.array(features.pop(label_name)) 
  history = model.fit(x=features, y=label, batch_size=batch_size,
                      epochs=epochs, shuffle=shuffle)
  
  # The list of epochs is stored separately from the rest of history.
  epochs = history.epoch

  # Isolate the classification metric for each epoch.
  hist = pd.DataFrame(history.history)

  return epochs, hist

def plot_curve(epochs, hist, list_of_metrics):
  """Plot a curve of one or more classification metrics vs. epoch."""  

  plt.figure()
  plt.xlabel("Epoch")
  plt.ylabel("Value")

  for m in list_of_metrics:
    x = hist[m]
    plt.plot(epochs[1:], x[1:], label=m)

  plt.legend()
````

# Train - test split

While random spliting is a good approach for many problems is not always the case. A frequent technique for online systems is to split the data by time, such that you would: collect 30 days, train 1-29, test 30. In this way the test task will match production task close.

But we need to know that label distribution over dates is not equal and at traning proportion of booked/notbooked will be smaller as the graphs show. Otherwise random spliting will help to bring the proportion equal.

![Split](files/model_notbooked.png "Split") Not booked to date ![Split](files/model_booked.png "Split") Booked to date

````python
df = pd.read_csv("./files/search_sessions.csv")
df.sort_values(by=['start'], inplace=True)
````

````python
# Spliting without shuffling to support onlime learning setting
train, test = train_test_split(df, test_size=0.2, shuffle=False)
````

````python
#
# For numerical stability in Deep Neural networs it is better
# to normalize numerical values with z-score normalisation
#

numeric_feature_names = [
    'adv_purchase', 'duration',
    'sess_time', 'query_freq_mean', 'searches',
    'distance', 'num_adults', 'num_children',
    
    # more sessions related features
    'origin_nuniq', 'destination_nuniq', 'num_adults_nuniq',
    'date_from_nuniq', 'date_to_nuniq', 'distance_mean',
    'adv_purchase_mean', 'duration_mean'
]

train_df_num = train[numeric_feature_names]
train_df_num_mean = train_df_num.mean()
train_df_num_std = train_df_num.std()
train_df_num_norm = (train_df_num - train_df_num_mean)/train_df_num_std

train.update(train_df_num_norm)

test_df_num = test[numeric_feature_names]
test_df_num_mean = test_df_num.mean()
test_df_num_std = test_df_num.std()
test_df_num_norm = (test_df_num - test_df_num_mean)/test_df_num_std

test.update(test_df_num_norm)
````

````python
categorical_feature_names = [
    'date_dow', 'origin', 'destination', 'date_from_dow', 'date_to_dow'
]

model_feature_columns = [
    tf.feature_column.numeric_column(feature_name) for feature_name in numeric_feature_names
]

# attach categorical features
model_feature_columns.extend([
    tf.feature_column.indicator_column(
        tf.feature_column.categorical_column_with_vocabulary_list(
            feature_name, vocabulary_list=df[feature_name].unique()))
    for feature_name in categorical_feature_names
])

# Convert the list of feature columns into a layer that will later be fed into the model. 
feature_layer = layers.DenseFeatures(model_feature_columns)
````

````python
# The following variables are the hyperparameters.
learning_rate = 0.002
epochs = 70
batch_size = 10000
classification_threshold = 0.35
label_name = 'booked'

# Here is the updated definition of METRICS:
METRICS = [
      tf.keras.metrics.BinaryAccuracy(name='accuracy', 
                                      threshold=classification_threshold),
      tf.keras.metrics.Precision(thresholds=classification_threshold,
                                 name='precision' 
                                 ),
      tf.keras.metrics.Recall(thresholds=classification_threshold,
                              name='recall'),
      tf.keras.metrics.Recall(thresholds=classification_threshold,
                              name='auc'),
]

# Establish the model's topography.
my_model = create_model(learning_rate, feature_layer, METRICS)

# Train the model on the training set.
epochs, hist = train_model(my_model, train, epochs, 
                           label_name, batch_size, shuffle=True)

# Plot metrics vs. epochs
list_of_metrics_to_plot = ['accuracy', 'precision', 'recall', 'auc'] 
plot_curve(epochs, hist, list_of_metrics_to_plot)
````

````python
features = {name:np.array(value) for name, value in test.items()}
label = np.array(features.pop(label_name))

my_model.evaluate(x = features, y = label, batch_size=batch_size)
````
